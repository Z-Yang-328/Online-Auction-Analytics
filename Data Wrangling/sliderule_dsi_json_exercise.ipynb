## JSON example, with string

+ demonstrates creation of normalized dataframes (tables) from nested json string
+ source: http://pandas.pydata.org/pandas-docs/stable/io.html#normalization

```python
# define json string
data = [{'state': 'Florida', 
         'shortname': 'FL',
         'info': {'governor': 'Rick Scott'},
         'counties': [{'name': 'Dade', 'population': 12345},
                      {'name': 'Broward', 'population': 40000},
                      {'name': 'Palm Beach', 'population': 60000}]},
        {'state': 'Ohio',
         'shortname': 'OH',
         'info': {'governor': 'John Kasich'},
         'counties': [{'name': 'Summit', 'population': 1234},
                      {'name': 'Cuyahoga', 'population': 1337}]}]
```

```python
# use normalization to create tables from nested element
json_normalize(data, 'counties')
```

```python
# further populate tables created from nested element
json_normalize(data, 'counties', ['state', 'shortname', ['info', 'governor']])
```

## JSON exercise

Using data in file 'data/world_bank_projects.json' and the techniques demonstrated above,
1. Find the 10 countries with most projects
2. Find the top 10 major project themes (using column 'mjtheme_namecode')
3. In 2. above you will notice that some entries have only the code and the name is missing. Create a dataframe with the missing names filled in.

```python
#Before we start, let's have a look at the dataset
world_bank = pd.read_json("data/world_bank_projects.json")
world_bank.head(2)
```

# Question 1: Find the 10 countries with most projects

1. In order to find 10 most-project countries, we need to group data by country names first. Since pd.groupby will not return dataframe directly, “reset_index()" will change pd.Series to pd.DataFrame (you can also use pd.Series.to_frame, but you'll also need to reset index after that);
2. Changing columns to make the results look better;
3. Sort by number of projects and display it from highest to lowest.

```python
#Create dataframe we need
countries = world_bank.groupby('countryshortname').size().reset_index()
#Change column names
countries.columns = ['Country name', 'Number of projects']
#Sort results
countries = countries.sort('Number of projects', ascending = False)
countries.head(10)
```

# Question 2: Find the top 10 major project themes (using column 'mjtheme_namecode')

```python
#Read and take a brief look at json data
#The 'jsons' here just like above json example file, please check above results for detail
#I will not display it here
jsons = json.load((open('data/world_bank_projects.json')))
```

```python
#Normalizing json data
world_bank_q2 = json_normalize(jsons, 'mjtheme_namecode',['countryname', 'countryshortname'])
#Change the results to dataframe and group by the project name
world_bank_q2_ans = pd.DataFrame({'name':world_bank_q2.name}).groupby(['name']).size()
world_bank_q2_ans
```

We can change the result to dataframe to make it look better

```python
world_bank_q2_df = world_bank_q2_ans.reset_index()
world_bank_q2_df.columns = ['Project name', 'Number of projects']
world_bank_q2_df
```

# Question 3 : Sort by number of projects and display it from highest to lowest.

1. In this dataframe, the missing values are represented by empty strings(''), which actually is not missing values in pandas dataframe. So the first thing we need to do is to tansfer empty strings to real null value;
2. Group by project name and show the results.

```python
#Change empty strings to None
world_bank_q2['name'].replace([''], [None], inplace = True)
#Change code column data type from string to int
world_bank_q2['code'] = world_bank_q2['code'].astype(int)
world_bank_q3_df = world_bank_q2.sort('code')
#Fill missing values, ffill means 'backward fill'
world_bank_q3_df.fillna(method = 'bfill', inplace = True)
world_bank_q3_df.head(10)
```

```python
#keep only name and code columns
world_bank_q3_ans = world_bank_q3_df.groupby(['code','name']).size().reset_index().sort(0, ascending = False)
world_bank_q3_ans.columns = ['Project code', 'Project name', 'Number of projects']
world_bank_q3_ans
```
