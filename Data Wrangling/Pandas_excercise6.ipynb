### Challenge: first combine these sales together into a single dataframe, then compute how much money consumers spent on each book in each currency.

```python
df1 = sales1.rename(columns={'Book title':'Title', 'Number sold':'Units sold', 
                             'Sales price':'List price', 'Royalty paid':'Royalty'})
df1['Currency'] = 'USD'
df1
```

```python
col = sales2.Title.str.extract(r'\((.*)\)')
sales2['Currency'] = col
sales2['Currency'] = sales2.Currency.fillna(method = 'bfill')
sales2
```

```python
df2 = sales2.dropna()
df2.replace(["The Bricklayer's Bible", "Pining for the Fisheries of Yore"], 
            ["The Bricklayer’s Bible", "Pining For The Fisheries of Yore"], inplace = True)
df2
```

```python
df = pd.concat([df1, df2])
df
```

```python
df['Total sales'] = df['List price'] * df['Units sold']
df.groupby(['Title', 'Currency'])[['Total sales']].sum()
```
